#NumPy(Array Operation)
1. Create a NumPy array of shape (5, 5) with random integers between 1 and 50.
o Find the row with the maximum sum.
o Replace all elements greater than the mean with 0.

```python
import numpy as np
arr=np.random.randint(1,51,size=(5,5))
print(arr)
row_wise_sum=np.sum(arr,axis=1)
print()
print("Sum row : ",row_wise_sum)
max_sum=np.max(row_wise_sum)
print("Max sum : ",max_sum)
mean=np.mean(arr)
arr[arr>mean]=0
print(arr)
```

##Pandas (Data Handling)
3. Load the dataset Iris data
o Display the first 10 rows and basic statistics.
o Filter all rows where sepal_length > median and species is "versicolor".
o Group the data by species and compute the mean petal_length.

```python
import pandas as pd
import seaborn as sns
iris=sns.load_dataset("iris")
iris.head(10)

```

```python
# Filter all rows where sepal_length > median and species is "versicolor".
sepal_length_median=iris["sepal_length"].median()
filtered_data=iris[(iris["sepal_length"]>sepal_length_median) & (iris["species"]=="versicolor")]
print(filtered_data)
```

```python
#Group the data by species and compute the mean petal_length.
grouped_data=iris.groupby("species")["petal_length"].mean()
df=pd.DataFrame(grouped_data)
df.columns=["Mean Petal Length"]
print(df)
```

4. Add a new column named petal_ratio = petal_length / petal_width.
o Find the top 5 rows with the highest petal_ratio.

```python
df=pd.DataFrame(iris)
df["petal_ratio"]=df["petal_length"]/df["petal_width"]
print(df)
```

##Data Visualization

5. Using Matplotlib, create:

o A histogram of sepal_length with 15 bins.




```python
import matplotlib.pyplot as plt
plt.hist(df["sepal_length"],bins=15)
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()
```

o A scatter plot of sepal_length vs petal_length, color-coded by species.

```python
# A scatter plot of sepal_length vs petal_length, color-coded by species.

sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette='Set1')
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()
```
